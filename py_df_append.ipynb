#### หรือ

```python
# append list to dataframe

rst = []
for i in [1,2]:
    tmp = globals()['df'+str(i)].copy()
    rst.append(tmp)

for n, table in enumerate(rst) :
    if n == 0 :
        df = table.copy()
    else :
        df = df.append(table, ignore_index=True)
        
        
df      
```
